```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 데이터 로드 (여기서는 df라는 DataFrame에 이미 데이터가 있다고 가정)
# df = pd.read_csv('your_data.csv')

# 고려할 변수들만 추출
data = df[['scaled_weighted_bd_high', 'Scaled Economic Indicator', 'scaled_pctrank', 'average_score']]

# 데이터 스케일링 (정규화)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# 엘보우 기법을 위한 KMeans 클러스터 수 찾기
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# 엘보우 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 데이터 준비 (예: df에서 필요한 컬럼 선택)
data = df[['scaled_weighted_bd_high', 'Scaled Economic Indicator', 'scaled_pctrank', 'average_score']]

# 데이터 정규화
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 군집 수 범위 설정
range_n_clusters = list(range(2, 11))  # 2에서 10까지 군집 수를 시도

# 평균 실루엣 점수 계산
average_silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_scaled)
    cluster_labels = kmeans.labels_
    
    # 실루엣 점수 계산
    silhouette_avg = silhouette_score(data_scaled, cluster_labels)
    average_silhouette_scores.append(silhouette_avg)
    print(f"군집 수: {n_clusters}, 평균 실루엣 점수: {silhouette_avg}")

# 군집 수에 따른 평균 실루엣 점수 시각화
plt.plot(range_n_clusters, average_silhouette_scores, marker='o')
plt.title("군집 수에 따른 평균 실루엣 점수")
plt.xlabel("군집 수")
plt.ylabel("평균 실루엣 점수")
plt.show()

# 최적의 군집 수 선택
best_n_clusters = range_n_clusters[average_silhouette_scores.index(max(average_silhouette_scores))]
print(f"최적의 군집 수: {best_n_clusters}")
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 갭 통계량 계산 함수
def gap_statistic(X, nrefs=10, maxClusters=10):
    # X: 데이터
    # nrefs: 무작위 데이터 세트 수
    # maxClusters: 최대 군집 수
    gaps = np.zeros(maxClusters-1)
    
    # 실제 군집화의 인에르시아(inertia) 계산
    for k in range(1, maxClusters):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        original_dispersion = kmeans.inertia_  # WCSS 계산

        # 무작위 데이터 세트 생성 후 군집화
        ref_dispersion = np.zeros(nrefs)
        for i in range(nrefs):
            random_data = np.random.random_sample(size=X.shape)  # 무작위 데이터
            kmeans.fit(random_data)
            ref_dispersion[i] = kmeans.inertia_

        # 갭 통계량 계산
        gaps[k-1] = np.log(np.mean(ref_dispersion)) - np.log(original_dispersion)
    
    return gaps

# 데이터 준비
data = df[['scaled_weighted_bd_high', 'Scaled Economic Indicator', 'scaled_pctrank', 'average_score']]

# 데이터 정규화
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 갭 통계량 계산
gaps = gap_statistic(data_scaled, nrefs=10, maxClusters=10)

# 갭 통계량 결과 시각화
plt.plot(range(1, 10), gaps, marker='o')
plt.title('갭 통계량 (Gap Statistic) 결과')
plt.xlabel('군집 수')
plt.ylabel('갭 통계량')
plt.grid(True)  # 그래프에 격자 추가
plt.show()

# 최적 군집 수 선택: 갭 통계량이 최대인 군집 수 선택
best_n_clusters_gap = np.argmax(gaps) + 1
print(f"갭 통계량에 따른 최적 군집 수: {best_n_clusters_gap}")
```

```python
from sklearn.metrics import davies_bouldin_score
import numpy as np

# 최적 군집 수 계산을 위한 함수 (데이비스-볼딘 지수)
def davies_bouldin(X, max_clusters=10):
    db_scores = np.zeros(max_clusters-1)
    
    for k in range(2, max_clusters+1):  # 군집 수 2부터 시작
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        db_score = davies_bouldin_score(X, kmeans.labels_)  # DB 지수 계산
        db_scores[k-2] = db_score
    
    return db_scores

# 데이터 준비 (정규화된 데이터 사용)
data = df[['scaled_weighted_bd_high', 'Scaled Economic Indicator', 'scaled_pctrank', 'average_score']]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 데이비스-볼딘 지수 계산
db_scores = davies_bouldin(data_scaled, max_clusters=10)

# 데이비스-볼딘 지수 그래프 시각화
plt.plot(range(2, 11), db_scores, marker='o')
plt.title('데이비스-볼딘 지수 (Davies-Bouldin Index)')
plt.xlabel('군집 수')
plt.ylabel('DB 지수')
plt.grid(True)
plt.show()

# 최적 군집 수 (최소값인 군집 수 선택)
best_n_clusters_db = np.argmin(db_scores) + 2  # 군집 수가 2부터 시작하므로
print(f"데이비스-볼딘 지수에 따른 최적 군집 수: {best_n_clusters_db}")
```

```python
from sklearn.metrics import calinski_harabasz_score

# Calinski-Harabasz Index 계산 함수
def calinski_harabasz(X, max_clusters=10):
    ch_scores = np.zeros(max_clusters-1)
    
    for k in range(2, max_clusters+1):  # 군집 수 2부터 시작
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        ch_score = calinski_harabasz_score(X, kmeans.labels_)  # CH 지수 계산
        ch_scores[k-2] = ch_score
    
    return ch_scores

# Calinski-Harabasz Index 계산
ch_scores = calinski_harabasz(data_scaled, max_clusters=10)

# Calinski-Harabasz Index 그래프 시각화
plt.plot(range(2, 11), ch_scores, marker='o')
plt.title('Calinski-Harabasz Index')
plt.xlabel('군집 수')
plt.ylabel('CH 지수')
plt.grid(True)
plt.show()

# 최적 군집 수 (최대값인 군집 수 선택)
best_n_clusters_ch = np.argmax(ch_scores) + 2  # 군집 수가 2부터 시작하므로
print(f"Calinski-Harabasz Index에 따른 최적 군집 수: {best_n_clusters_ch}")
```
